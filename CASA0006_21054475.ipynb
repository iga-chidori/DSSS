# 2.Literature Review

For people living in modern society, traffic accidents on the road are always a huge threat to life safety. According to [WHO](#WHO-2021) ([2021](#WHO-2021)), road traffic accidents are responsible for approximately 1.3 million deaths worldwide every year, while also causing a loss of 3% of gross domestic product in a majority of countries. In Sustainable Development Goal (SDG) targets 11.2, the objective that aims to provide all people in the world safe, accessible, sustainable and affordable transport systems by 2030 was been made ([WHO, 2017](#WHO-2017)). In an ideal scenario, the prediction system of the traffic accident severity is able to estimate potential consequences of ongoing traffic accidents and provide critical information to emergency responders in order to facilitate timely response to traffic safety management strategies ([Gan et al., 2020](#Gan-et-al)). By such approach, it is possible to strongly contribute to the achievement of SDG targets 11.2.

However, the development of traffic accidents is an extremely complex process covering a myriad of interactions of characteristic variables and there are often stochastic perturbations in the traffic accidents. A traffic accident usually covers a range of factors such as human, vehicles, weather, road conditions, and unexpected situations, and given such characteristics, predicting the severity of a traffic accident is considered a highly challengeable task. In the past, a great deal of work using statistical models has been done to make predictions of traffic accidents severity. But in recent years, with the emergence of more flexible machine learning models, more attention has been moved to the application of learning models in order to handle such complex predictions that contain dozens of features. In the research from [Moghadda et al.](#Moghaddam-et-al) ([2011](#Moghaddam-et-al)), the model of Artificial Neural Networks was utilized to achieve the predictions of traffic accident severity and capture important feature that are relevant to accidents and highways. [Yan and Shen](#Yan-and-Shen) ([2022](#Yan-and-Shen)) integrated Random Forest and Bayesian optimization as a hybrid model with optimisation of hyperparameters to accomplish the estimation of traffic accident severity. [Taamneh et al.](#Taamneh-et-al) ([2017](#Taamneh-et-al)) built several models including Decision Tree, Multilayer Perceptron and Naive Bayes and compared the performance of models in predicting traffic accident severity. Machine learning models have unique advantages in the aspect of accuracy and exemption that assumptions of the relationship between explanatory variables and response variables are not required compared to traditional statistical models ([Gan et al., 2020](#Gan-et-al)).

## 4.1.Data Loading

The datasets that used in this study are provided by [*Department for Transport*](https://data.gov.uk/dataset/cb7ae6f0-4be6-4935-9277-47e5ce24a11f/road-safety-data). The license for the datasets is the [*Open Government Licence*](https://www.nationalarchives.gov.uk/doc/open-government-licence/version/3/). 

This project contained two types of raw datasets covering **accident details** and **vehicle details** for traffic accidents from 2019 to 2020 in UK. The documented traffic road safety datasets of 2019 and 2020 were validated by the data provider and all the categorical and ordinal variables were already encoded rather than using textual descriptions.

```python
accidents_urls = [
    'https://data.dft.gov.uk/road-accidents-safety-data/dft-road-casualty-statistics-accident-2020.csv',
    'https://data.dft.gov.uk/road-accidents-safety-data/dft-road-casualty-statistics-accident-2019.csv',
]

vehicles_urls = [
    'https://data.dft.gov.uk/road-accidents-safety-data/dft-road-casualty-statistics-vehicle-2020.csv',
    'https://data.dft.gov.uk/road-accidents-safety-data/dft-road-casualty-statistics-vehicle-2019.csv',
]

accidents_csv = []
vehicles_csv = []

for url in accidents_urls:
    accidents_csv.append(pd.read_csv(url))
    
for url in vehicles_urls:
    vehicles_csv.append(pd.read_csv(url))
    
accidents = pd.concat(accidents_csv, ignore_index=True)
vehicles = pd.concat(vehicles_csv, ignore_index=True)
accidents['accident_index'] = accidents['accident_index'].astype(str)
vehicles['accident_index'] = vehicles['accident_index'].astype(str)
```

For the reason that traffic accident may involve multiple vehicles, the datasets could be integrated into one via `merge` according to `accident_index`.

```python
print('rows of accidents:', len(accidents), ';',
      'rows of vehicles:', len(vehicles))
```

```python
# Left join accidents to vehicles
raw = pd.merge(vehicles, accidents,
              how = 'left',
              on = 'accident_index')
```

## 4.2.Data Cleaning

In order to clean and collate the integrated datasets, it is necessary to look up [the metadata of the original datasets](https://data.dft.gov.uk/road-accidents-safety-data/Road-Safety-Open-Dataset-Data-Guide.xlsx).
The process of data cleaning was listed below:
1. Extract temporal features from `data` and `time`.
2. Create mapping between encoded characters and labels for response variable `accident_severity` and refer back to the labels.
3. Transform missing fields to `NaNs` and examine the presence of missing data.
4. Drop features that containing over 40% `NaNs`; weakly related to the predictions; and is highly dependent on other explanatory variables. 
5. Process and manage the remaining fields of `NaNs` by filling `-1` or `'-1'` and respectively create a new dummy variable that indicating the presence of missing data in features correspondingly.
6. Transform and validate the correct datatype of each feature.

```python
raw = raw.rename(columns={'accident_year_x':'accident_year'})
raw['hour'] = raw['time'].map(lambda x:x.split(':')[0]).astype(int)
raw['day'] = raw['date'].map(lambda x:x.split('/')[0]).astype(int)
raw['month'] = raw['date'].map(lambda x:x.split('/')[1]).astype(int)
```

```python
index_to_label = {
  1: 'fatal',
  2: 'serious',
  3: 'slight',
}
raw['accident_severity'] = raw['accident_severity'].map(index_to_label)
```

```python
df = raw.replace({-1, '-1'},np.nan)
```

```python
total = df.isnull().sum().sort_values(ascending=False)
percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)

missing_data = pd.concat(
    [total, percent],
    axis=1,
    keys=['Total', 'Percent'])
missing_data.head(15)
```

Variables that are less related with the prediction of accident severity and have too many `NaNs` (over 40%) should be dropped.

```python
df.columns
```

```python
# Redundant, irrelevant, or repetitive features will be manually dropped
to_drop = ['accident_reference_x', 'vehicle_reference','accident_year_y', 'accident_reference_y',
           'location_easting_osgr', 'location_northing_osgr', 'vehicle_direction_from', 'vehicle_direction_to',
           'police_force','local_authority_district', 'hit_object_in_carriageway',
           'date', 'time', 'local_authority_highway','hit_object_off_carriageway',
           'driver_imd_decile', 'local_authority_ons_district', 'did_police_officer_attend_scene_of_accident',
           'lsoa_of_accident_location', 'age_band_of_driver',
           'generic_make_model', 'junction_control']
df2 = df.drop(to_drop, axis = 1)
```

The completeness of coordinates should be checked and any records that lack of location reference (`longitude` and `latitude`) need to be dropped. Because for those missing data, it was not possible to apply <i>fillna</i> with -1.

```python
print(len(df2[df2['longitude'].isnull()]), len(df2[df2['latitude'].isnull()]))
```

```python
df_withnull = df2.drop(df2[df2['longitude'].isnull()].index, axis = 0)
```

The features with missing data should be listed and checked again. The missing data in those features were filled as `-1` and `'-1'` (depending on the data type), and dummy variables `[Feature_Name]_G` were set up for continuous data respectively. The data types of variables should be carefully examined according to the [Metadata](https://data.dft.gov.uk/road-accidents-safety-data/Road-Safety-Open-Dataset-Data-Guide.xlsx) and transformed manually.

```python
total_2 = df_withnull.isnull().sum().sort_values(ascending=False)
percent_2 = (df_withnull.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
missing_data_2 = pd.concat([total_2, percent_2], axis=1, keys=['Total', 'Percent'])
Features_Missing = missing_data_2[missing_data_2['Total'] > 0].index.tolist()
```

```python
data_processed = df_withnull.copy()
for feature in Features_Missing:
    data_processed[feature] = df_withnull[feature].fillna(-1)
data_processed = data_processed.set_index('accident_index')
```

```python
to_int = ['accident_year', 'age_of_driver', 'engine_capacity_cc', 'age_of_vehicle', 'speed_limit',
          'number_of_vehicles', 'number_of_casualties', 'hour', 'day', 'month', 'day_of_week',
          'first_road_number', 'second_road_number', ]

to_label = ['vehicle_type', 'towing_and_articulation', 'vehicle_manoeuvre', 'vehicle_location_restricted_lane', 'junction_location', 
            'vehicle_leaving_carriageway', 'vehicle_left_hand_drive', 'journey_purpose_of_driver', 'sex_of_driver', 'propulsion_code', 
            'driver_home_area_type', 'first_road_class', 'road_type', 'skidding_and_overturning',
            'junction_detail', 'second_road_class', 'pedestrian_crossing_human_control','first_point_of_impact',
            'pedestrian_crossing_physical_facilities', 'light_conditions', 'weather_conditions', 'road_surface_conditions',
            'special_conditions_at_site', 'carriageway_hazards', 'urban_or_rural_area', 'trunk_road_flag']

data_processed[to_int] = data_processed[to_int].astype(int)
data_processed[to_label] = data_processed[to_label].astype(int).astype(str)
# Transform features into correct data types
```

```python
int_missing = list(
    set(Features_Missing).intersection(
        set(to_int)
    )
)

# Extract features in int type that have NaNs
for feature in int_missing:
    data_processed[feature+'_G'] = (data_processed[feature] == -1).astype(int)
# Create a dummy variable named '[feature_name]_G' indicating the presence of NaNs
```

```python
# To examine the correlation matrix of continuous variables
corr = data_processed.corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(7, 5))
    ax = sns.heatmap(corr, mask=mask, vmax=.3, square=True)
```

We could see that, in the processed dataset, most features are independent. Features that highly correlated are occurred in features related to location and vehicle conditions, and the dummy variables and for indicating `NaNs`.

### 5.3.4.Hyperparameter Tunning for RF

As the size of dataset was extremely large, the approach of holdout validation using predefined trainig and development set was considered to tune the RF model hyperparameters in order to optimise model performance. 

The tunning of hyperparameters included the tunning all hyperparameters (`n_estimators`, `max-depth`, `max_features`, `min_samples_leaf`, `min_samples_split`, and etc.) for optimising the model performance. For reducing the running time, only one hyperparameter - `n_estimators` was selected to optimise the model and the selection of values of this hyperparameter was set to be among 50, 75, 100, 150, 200, 400. Due to the large size of dataset, selection of `n_estimators` was controlled in a small range with restricted numbers. 

```python
# merge training and validation data, as required by GridSearchCV
d_train_val = np.concatenate(
    (train_d, val_d),
    axis=0)

lab_train_val = np.concatenate(
    (train_lab, val_lab),
    axis=0)

# the index list of training and validation data in the merged dataset
ind_train = list(range(train_d.shape[0]))
ind_val = list(range(train_d.shape[0], d_train_val.shape[0]))
```

```python
%%time
# values of n_estimators
parameters = {
    'n_estimators':[50,75,100,150,200,400]
}

rf = RandomForestClassifier(random_state = random_state)

# CV: An iterable yielding (train, test) splits as arrays of indices.
clf = model_selection.GridSearchCV(
    rf,
    parameters,
    cv=[(ind_train, ind_val)]
)

clf.fit(d_train_val, lab_train_val)

# we can query the best parameter value and its accuracy score
print ("The best parameter value is: ")
print (clf.best_params_)
print ("The best score is: ")
print (clf.best_score_)
```

Through hyperparameter tuning, the best parameter of `n_estimators` remains to be 100 (same as default), which leads to no much improvement on RF model.

```python
%%time
rf_final = RandomForestClassifier(
    n_estimators=clf.best_params_['n_estimators'],
    random_state=random_state)
rf_final.fit(train_d,train_lab)
```

```python
print("The accuracy of final RF classifier on the train data is:{}".format(rf_final.score(train_d, train_lab)))
print("The accuracy of final RF classifier on the dev data is:{}".format(rf_final.score(val_d, val_lab)))
```

## 6.2.Feature Importance of RF Model

From RF model, interpretable results of relative importance could be obtained. The sum of partial importance of all features is 1. To explicitly examine the relative importance of all selected features, the feature importance of every categorial variable is summed.

We can identify important influential factors for traffic accident severity through examining relative importance. 
Such results would provide insights to figure out the key feature that relate to the severity of traffic accident and may help to mitigate the severity of traffic accident consequences and eventually contribute to the sustainable development of transportation.

```python
importance = forest_clf.feature_importances_
# Obtain the partial importance of each feature

Feature_import = pd.DataFrame(
    {
     'Feature':feature,
     'Score': importance.tolist()
     }
)
# Create a new dataframe of feature importance.
```

```python
Feature_import['Test'] = Feature_import['Feature'].str.contains('=')
# To test whether this feature is a dummy variable derived from categorial variable
Feature_import['Group'] = ''
# Create a new column of 'Group'

for i in range(len(Feature_import)):
    if Feature_import['Test'][i] == True:
        Feature_import['Group'][i] = Feature_import['Feature'][i].split('=')[0]
    else: 
        Feature_import['Group'][i] = Feature_import['Feature'][i]
# For every dummy variable derived originally from categorial variable,
# this iteration extracts their the original feature name

Feature_import_sum = Feature_import.groupby(
    by='Group'
).sum().drop(
    'Test',
    axis = 1
).reset_index()
# According to name of feature, sum up the partial importance of each feature

Feature_import_sum.sort_values(
    by='Score',
    ascending=False
).round(4).head(20)
# Rank it
```

```python
f, ax = plt.subplots(figsize=(7, 12))

head_feature = Feature_import_sum.sort_values(
    by='Score',
    ascending=False
).round(4)

ax = sns.barplot(
    x="Score",
    y="Group",
    data=head_feature,
    color = 'r')

ax.set_ylabel('Feature')
```

Due to the huge number of transformed input features with dummy variables, here we merge them again into the initial feature categories and then show the partial importance of all features.

From the ranked barchart of feature importance, following findings could be addressed: Due to large number of input features, partial importance of each feature is relatively small and single feature that maximally contribute to the prediction of traffic severity only accounts for 6%; Spatial location (`latitude` and `longitude`) where accidents occurred are most critical factor related to the severity of traffic accident; Temporal features of `day`, `hour`, `month` and `day_of_week` are the second important to predict; Next, `age_of_driver`, `vehicle_manoeuvre` when accident happened and conditions of vehicles including `age_of_vehicle` and `engine_capacity` are the following most influential variables.

## 6.3.Limitations

Restricted by the availability of historic weather data in the UK, daily weather data from nearby climate stations are not added to each traffic event. Also, limited by the vague road information in the provided dataset, it is not convenient to merge into the traffic flow dataset.

In this research, pre-process operators like normalization and standardization were not applied to the data though such pre-processing is likely to yield better results empirically. However, since the raw dataset used in this project was highly imbalanced, basic classifier cannot effectively improve the performance of the models, no matter how to add or subtract features during selection, adjust the data types, tune the split of dataset and optimise hyperparameters.

Additionally, when evaluating the final performance of the models, this project only considered the mostly widely used metrics for classification. According to [Tanha et al.](#Tanha-et-al) ([2020](#Tanha-et-al)), other suitable evaluation metrics that particularly built for MCC in imbalanced dataset like *Kappa*, *Weighted Accuracy* could be employed. Meanwhile, *Receiver Operating Characteristic (ROC) curve* and *Precision-recall curves (PR curves)* are also recommended for considering imbalanced issues ([Branco et al., 2017](#Branco-et-al)).

## 6.4.Future Works

As what has been stated in limitations, more work in the selection of representative features and adding features from daily weather data and flow data could be done in the next stage of investigation in the future. To mitigate the impact of imbalanced datasets on MCC, techniques of undersampling of the majority class, oversampling of the minority class, or integrated applications of these two sampling approaches could be conducted before model implementation  ([Liu et al., 2009](#Liu-et-al); [Mohammed et al., 2020](#Mohammed-et-al)). Though, by sampling techniques new datasets will be generated synthetically, it might adaptively increase the performance of learning models for imbalanced issues in MCC and accomplish the tasks for accurate predictions of traffic severity.

More advanced learning algorithms such as *AdaC1, AdaC2, AdaC3, SMOTEBoost, RUSBoost, MEBoost, AdaCost, SMOTE* and etc. are good adaption algorithms for MCC to solve imbalanced problems ([Tanha et al., 2020](#Tanha-et-al)). Future comparisons of the applications of these algorithms could be considered.
